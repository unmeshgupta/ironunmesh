```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Send a request to the Billboard Hot 100 page
url = "https://www.billboard.com/charts/hot-100"
response = requests.get(url)
```

```python

# Use BeautifulSoup to extract the data from the HTML
soup = BeautifulSoup(response.content, 'html.parser')
songs = soup.find_all('li', class_='chart-list__element')
```

```python
print(soup.prettify())
```

```python
# Store the data in a pandas DataFrame
df = pd.DataFrame(columns=['Title', 'Artist'])

for song in songs:
    title = song.find('span', class_='chart-element__information__song text--truncate color--primary').get_text()
    artist = song.find('span', class_='chart-element__information__artist text--truncate color--secondary').get_text()
    df = df.append({'Title': title, 'Artist': artist}, ignore_index=True)

```

```python
# Export the data to a CSV file
df.to_csv('hot100.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Request to the specific Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_best-selling_singles"
response = requests.get(url)

# Use BeautifulSoup to extract the data via HTML
soup = BeautifulSoup(response.content, 'html.parser')
songs = soup.find_all('table', class_='wikitable sortable')[0].find_all('tr')[1:]

# Store the data in a pd DataFrame
df = pd.DataFrame(columns=['Artist', 'Song', 'Year'])

for song in songs:
    cells = song.find_all('td')
    artist = cells[0].get_text('td')
    song = cells[1].get_text()
    year = cells[2].get_text()
    df = df.append({'Artist': artist, 'Song': song, 'Year': year}, ignore_index=True)

# Export the data to a CSV file
#df.to_csv('popular_songs.csv', index=False)
```
